```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
from dataclasses import dataclass
import json


@dataclass
class Item:
    sport_league: str = ''
    event_date_utc: str = ''
    team1: str = ''
    team2: str = ''
    pitcher: str = ''
    period: str = ''
    line_type: str = ''
    price: str = ''
    side: str = ''
    team: str = ''
    spread: float = 0.0


driver = webdriver.Chrome()


driver.get("https://veri.bet/simulator")


driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)


access_button = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//a[contains(text(),'Access Betting Simulator')]")))


driver.execute_script("arguments[0].click();", access_button)


import time
time.sleep(5)


page_source = driver.page_source


driver.quit()


soup = BeautifulSoup(page_source, 'html.parser')


items = []


for row in soup.find_all('tr', class_='match-row'):
    cols = row.find_all('td')
    item = Item(
        sport_league=cols[0].text.strip(),
        event_date_utc=cols[1].text.strip(),
        team1=cols[2].text.strip(),
        team2=cols[3].text.strip(),
        pitcher=cols[4].text.strip(),
        period=cols[5].text.strip(),
        line_type=cols[6].text.strip(),
        price=cols[7].text.strip(),
        side=cols[8].text.strip(),
        team=cols[9].text.strip(),
        spread=float(cols[10].text.strip())
    )
    items.append(item)

# Exiba os dados como JSON
print(json.dumps([item.__dict__ for item in items], indent=2))
```
